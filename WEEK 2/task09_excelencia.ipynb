```python
import pandas as pd
df = pd.read_csv("Downloads/sales_data.csv")
```

```python
df.head()
```

```python
pt = pd.pivot_table(df, values = "Revenue", index = "Region", columns = "Category", aggfunc = "sum", fill_value = 0)
print("Revenue by Region × Category")
pt
```

```python
pt1 = pd.pivot_table(df, values = "Revenue", index = "Region", columns = ["Category", "Product"], fill_value = 0)
print("Generl pivot table")
pt1
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.dtypes
```

```python
df["Month"] = df["Date"].dt.to_period("M")
month = df["Month"]
```

```python
monthly = df.groupby("Month")["Revenue"].sum().reset_index()
monthly
```

```python
#monthly = df.groupby("Month")["Revenue"].sum().reset_index()
monthly["Running_Total"] = monthly["Revenue"].cumsum()
monthly["Running_Total"]
```

```python
print("Montly Revenue")
print(monthly)
```

```python
df["Region_Total"] = df.groupby("Region")["Revenue"].transform("sum")
df["Percent_of_Region"] = (df["Revenue"] / df["Region_Total"] * 100).round(2)
print("Percentage of Region Total")
print(df[["Region", "Product", "Revenue", "Percent_of_Region"]])
```

```python
monthly_region = (df.groupby(["Region", "Month"])["Revenue"].sum().reset_index())

monthly_region = monthly_region.sort_values(["Region", "Month"])

monthly_region["Rolling_Avg"] = (monthly_region.groupby("Region")["Revenue"]
                                 .rolling(window=3, min_periods=1).mean()
                                 .reset_index(level=0, drop=True))
```

```python
print("3-Month Rolling Average Revenue per Region")
print(monthly_region)
```
