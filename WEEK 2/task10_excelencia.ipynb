```python
import pandas as pd
order = pd.read_csv("Downloads/orders.csv")
cust = pd.read_csv("Downloads/customers.csv")
prdt = pd.read_csv("Downloads/products.csv")
mr = pd.read_csv("Downloads/monthly_revenue.csv")
```

```python
df1, df2, df3, df4 = order, cust, prdt, mr 
```

```python
orders_products = pd.merge(
    df1,
    df3,
    on="ProductID",
    how="inner"
)
orders_products
```

```python
final_df = pd.merge(
    orders_products,
    df2,
    on="CustomerID",
    how="inner"
)

print(final_df.head())

print("Shape :", final_df.shape)
```

```python
#Validate No Rows Lost on Inner Join

print("Original Orders :", len(df1))
print("Rows After Inner Join :", len(final_df))

rows_lost = len(df1) - len(final_df)

print("Rows Lost :", rows_lost)
```

```python
long_df =  df4.melt(
    id_vars = "Month",
    var_name = "Region",
    value_name = "Revenue",
    ignore_index = True
)
long_df
```

```python
#Orders with No Matching Customer

left_join = pd.merge(
    df1,
    df2,
    on="CustomerID",
    how="left",
    indicator=True
)

left_join
```

```python
missing_customers = left_join[left_join["_merge"] == "left_only"]
missing_customers.count()
```

```python
print("Orders Without Matching Customer :")
print(missing_customers.head())

print("Total Missing Customers :", len(missing_customers))
```
