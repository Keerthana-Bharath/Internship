```python
import pandas as pd
df = pd.read_csv("Downloads/orders_strings_dates.csv")
df.head()
```

```python
#Standardise product_name

df["product_name"] = (df["product_name"].str.strip()
                     .str.title()
                     .str.replace(r"[^A-Za-z0-9]", "",regex = True))
```

```python
df.head()
```

```python
#Extract region code using Regex

df["region_code"] = df["ship_address"].str.extract(r"\((.*?)\)")
df.head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df["days_to_ship"] = (df["ship_date"] - df["order_date"]).dt.days
df.head()
```

```python
#lag Weekend Shipments

df["shipped_on_weekend"] = (df["ship_date"].dt.dayofweek >= 5)
```
