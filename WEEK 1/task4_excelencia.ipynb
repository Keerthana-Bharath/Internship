```python
import pandas as pd
df = pd.read_csv("Downloads/sales_raw.csv")
df
```

```python
print(df.info())
```

```python
df.columns = df.columns.str.strip()
```

```python
df
```

```python
print(df.isna().sum())
```

```python
df["Quantity"] = df["Quantity"].fillna(0)
```

```python
df
```

```python
df["Price"] = df["Price"].fillna(0)
```

```python
df
```

```python
print(df.isna().sum())
```

```python
df["Quantity"] = df["Quantity"].astype(int)

df["Price"] = df["Price"].astype(float)
```

```python
df
```

```python
from pathlib import Path

clean_file = Path("sales_clean.csv")

df.to_csv(
    clean_file,
    index=False
)
```

```python
clean_file
```

```python
import json

summary = {

    "Total Rows": len(df),

    "Total Quantity": int(df["Quantity"].sum()),

    "Total Sales": float(
        (df["Quantity"] * df["Price"]).sum()
    ),

    "Average Price": float(
        df["Price"].mean()
    )
}
```

```python
summary
```

```python
summary_file = Path("sales_summary.json")

with open(summary_file, "w") as file:

    json.dump(
        summary,
        file,
        indent=4
    )
```
