```python
def clean_price(val):
    print(val)
val = input("Enter the price").strip('$')
clean_price(val)

try:
    FP = float(val)
except ValueError as e:
    print(e)
except TypeError as e:
    print(e)
finally:
    print("The float value is:",FP)
```

```python
def categorize_sales(amt):
    if amt < 2000:
        print("low")
    elif amt >= 2000 and amt <= 5000:
        print("medium")
    else:
        print("high")
amt = int(input("Enter the amt"))     
categorize_sales(amt)
```

```python
def top_n(records):
    sorted_list = sorted(records, key = lambda x : x, reverse = True)
    print("The sorted list is: \n",sorted_list)
l = [10,20,50,30,40]
top_n(l)
```
