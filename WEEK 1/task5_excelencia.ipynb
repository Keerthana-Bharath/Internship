```python
import pandas as pd
class DataPipeline:
    
    def __init__(self, filename):
        self.filename = filename
        self.data = None

    def load(self):

        self.data = pd.read_csv(self.filename)
        print("Data Loaded Successfully")
        print("Rows :", len(self))
        return self

    def clean(self):

        self.data.columns = self.data.columns.str.strip() #Remove spaces
        self.data = self.data.drop_duplicates() #Remove duplicate rows
        self.data = self.data.dropna() #Remove missing values
        print("Data Cleaned Successfully")
        print("Rows :", len(self))
        return self

    def export(self, output_file):

        self.data.to_csv(output_file, index=False)
        print("Data Exported Successfully")
        print("Rows :", len(self))
        return self

    def __len__(self):
        
        if self.data is None:
            return 0
        else:
            return len(self.data)

    def __repr__(self):

        return f"DataPipeline('{self.filename}')"

#Create pipeline obj
pipeline = DataPipeline("sales_raw.csv")

#Method Chaining
pipeline.load().clean().export("output.csv")

#Display the object
print(pipeline)

#Display the number of rows
print("Final Row Count:", len(pipeline))
```
