```python
import numpy as np

sales = np.random.normal(
    loc=5000, #loc = mean; provides vals around the mean val
    scale=1000, #scale = std dev; between 4000 to 6000
    size=1000 #generates 1000 vals
)
```

```python
mean_sales = np.mean(sales)
mean_sales
```

```python
std_sales = np.std(sales)
std_sales
```

```python
p25 = np.percentile(sales, 25)
p25
```

```python
p75 = np.percentile(sales, 75)
p75
```

```python
days_above_avg = np.sum(sales > mean_sales)
days_above_avg
```

```python
hist, bins = np.histogram(sales,bins=10)

print("Histogram Counts:")
print(hist)

print("\nBin Edges:")
print(bins)
```
